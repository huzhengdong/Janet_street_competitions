```python
def data_preprocessing(jane_street_data, test_size=0.2, random_state=42, na_values='delete', if_log = True, if_standard=True, if_process_labels=False):
    #Three options for na_values: 'delete', 'zero', or 'mean'.
    #if_log will not log normalize the data except for data and feature_0
    #if_process_labels will log and standardize the labels
    import math
    from sklearn.model_selection import train_test_split
    
    print('Now in preprocessing...')
    
    
    #Process NA values
    if na_values == 'delete':
        jane_street_data.dropna(inplace=True)
        
    elif na_values == 'zero':
        jane_street_data.fillna(0, inplace=True)
        
    elif na_values == 'mean':
        for column in list(jane_street_data.columns[jane_street_data.isnull().sum() > 0]):
            mean_val = jane_street_data[column].mean()
            jane_street_data[column].fillna(mean_val, inplace=True)
    
    else: print('Error: unexpected na_values input')
    
    
    #Logrithm normalization
    def log_norm_helper(x,log_bias=1):
        if x>=0:
            return math.log(x+log_bias)
        else:
            return -math.log(abs(x)+log_bias)
    
    if if_log:
        
        for column in jane_street_data.columns.tolist():
            
            if (not(column in ['date', 'feature_0','ts_id' ])):
                if if_process_labels:
                    jane_street_data[column] = jane_street_data[column].apply(log_norm_helper)
                    
                elif (not(column in ['resp_1', 'resp_2', 'resp_3', 'resp_4', 'resp'])):
                        jane_street_data[column] = jane_street_data[column].apply(log_norm_helper)
    
    
    #Split training set and evaluation set
    labels = jane_street_data[['resp_1', 'resp_2', 'resp_3', 'resp_4', 'resp']].copy()
    features = jane_street_data.drop(['resp_1', 'resp_2', 'resp_3', 'resp_4', 'resp'],axis=1)
    features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=test_size, random_state=random_state)
    
    
    #Standardization
    def standard_helper(x, mean, std):
        return (x-mean)/std if (std!=0) else x-mean       
    
    if if_standard:
        feature_mean = features_train.mean()
        feature_std = features_train.std()
        
        for column in features_train.columns.tolist():
            features_train.loc[:,column] = features_train[column].apply(standard_helper, mean=feature_mean[column], std=feature_std[column])
            features_test.loc[:,column] = features_test[column].apply(standard_helper, mean=feature_mean[column], std=feature_std[column])
            
        if if_process_labels:
            label_mean = labels_train.mean()
            label_std = labels_train.std()
            
            for column in labels_train.columns.tolist():
                labels_train[column] = labels_train[column].apply(standard_helper, mean=label_mean[column], std=label_std[column])
                labels_test[column] = labels_test[column].apply(standard_helper, mean=label_mean[column], std=label_std[column])

    print('Done in preprocessing!')

                
    return  features_train, features_test, labels_train, labels_test
```

```python
def train_evaluate(features_train, features_test, labels_train, labels_test, model='lasso',alpha=1):
    #Three options for model: 'linear', 'ridge', 'lasso'
    
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LinearRegression, Lasso,Ridge
    from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
    import numpy as np

    
    if model=='lasso':
        lin_reg = Lasso(alpha=alpha)
    elif model =='Ridge':
        lin_reg = Ridge(alpha=alpha)
    elif model =='linear':
        lin_reg = LinearRegression()
    else: print('Error: unexpected model input')
        
    
    print('Now in trianing...')
    lin_reg.fit(features_train, labels_train)
    print('Done in trianing!')
    
    print('Now in prediction...')
    predictions = lin_reg.predict(features_test)
    print('Done in prediction!')

        
    rmse = np.sqrt(mean_squared_error(np.array(labels_test),predictions))
    mae = mean_absolute_error(np.array(labels_test),predictions)
    r_2 = r2_score(np.array(labels_test),predictions)
    
    return rmse, mae, r_2
```

```python
sample = jane_data[jane_data.date <= 10]
```

```python
features_train, features_test, labels_train, labels_test = data_preprocessing(sample)
rmse, mae, r_2 = train_evaluate(features_train, features_test, labels_train, labels_test)
print('rmse:', rmse)
print('mae:', mae)
print('r_2:',r_2)
```
